```python
class Walker:
    x: int = 0
    y: int = 0
    walk_map: dict = {
        "R" :  1,
        "D" : -1,
        "L" : -1,
        "U" :  1
    }

    def move(self, direction: str):
        if direction in "RL":
            self.x += self.walk_map[direction]
        else:
            self.y += self.walk_map[direction]


class Head(Walker):
    pass


class Tail(Walker):
    pinned_head = None

    @classmethod
    def pin_to_head(cls, head):
        tail = Tail()
        tail.pinned_head = head
        return tail

    def chase_head(self, head: Head):
        x_diff = head.x - self.x
        y_diff = head.y - self.y
        x_dist = abs(x_diff)
        y_dist = abs(y_diff)
        if x_dist == 0 and y_dist > 1:
            self.y += y_diff//y_dist
        elif x_dist > 1 and y_dist == 0:
            self.x += x_diff//x_dist
        elif x_dist > 1 or y_dist > 1:
            self.x += x_diff//x_dist
            self.y += y_diff//y_dist
        
    def pin_chase(self):
        self.chase_head(self.pinned_head)
```

```python
def chase(directions: str) -> int:
    head = Head()
    tail = Tail()
    uniq_coords = set()
    for direction in directions:
        head.move(direction)
        tail.chase_head(head)
        uniq_coords.add((tail.x, tail.y))
    return len(uniq_coords)
```

```python
chase(data)
```

```python
def chase_pin(directions: str, rope_length: int) -> int:
    uniq_coords = set()
    head = Head()
    rope = [head]
    for i in range(rope_length-1):
        rope.append(Tail.pin_to_head(rope[-1]))

    print(len(rope))
    
    for direction in directions:
        head.move(direction)
        for knot in rope[1:]:
            knot.pin_chase()
        uniq_coords.add((rope[-1].x, rope[-1].y))
    
    return len(uniq_coords)
```

```python
chase_pin(data, 10)
```
